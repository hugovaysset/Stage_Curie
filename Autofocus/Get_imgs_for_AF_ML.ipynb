# Parse images from the imgs_for_AF_ML directories

```python
import os
import pandas as pd
import shutil
```

```python
# parse and copy the images
nas_dir = "N:/Tests"
target_dir = "D:/Hugo/Autofocus/Tests_AF_ML"

for i, expe in enumerate(os.listdir(nas_dir)):
    
    if "imgs_for_AF_ML" in os.listdir(f"{nas_dir}/{expe}"):
        
        for j, img in enumerate(os.listdir(f"{nas_dir}/{expe}/imgs_for_AF_ML")):
            if len(str(i)) < 2 and len(str(j)) < 2:
                shutil.copy(f"{nas_dir}/{expe}/imgs_for_AF_ML/{img}", f"{target_dir}/0{i}_0{j}_{expe}_{img}")
            elif len(str(i)) < 2 and len(str(j)) == 2:
                shutil.copy(f"{nas_dir}/{expe}/imgs_for_AF_ML/{img}", f"{target_dir}/0{i}_{j}_{expe}_{img}")
            elif len(str(i)) == 2 and len(str(j)) < 2:
                shutil.copy(f"{nas_dir}/{expe}/imgs_for_AF_ML/{img}", f"{target_dir}/{i}_0{j}_{expe}_{img}")
            else:
                shutil.copy(f"{nas_dir}/{expe}/imgs_for_AF_ML/{img}", f"{target_dir}/{i}_{j}_{expe}_{img}")
```

```python
# create a table with the images name
table = []

for img in os.listdir("D:/Hugo/Autofocus/Tests_AF_ML/"):
    print(img)
    table.append({"file_name": img, "z": -1})
    
table = pd.DataFrame(table)
table.to_csv("D:/Hugo/Autofocus/Tests_imgs_for_AF_ML.csv", sep=",")
```
