# Visual monitoring of the anaphase prediction

Here we use the generated measures and predictions of the TTA to generate a video on which the contours of a cell are coloured given the predicted TTA

```python
# %matplotlib inline
import cv2
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy
import seaborn as sns
from skimage import measure # to get contours from masks
import sklearn

import napari
```

# Load the data

```python
os.chdir("D:/Hugo/Anaphase/Structured/H449.1")

# get images
bf = imageio.volread("f0_BF.tif")
rfp  = imageio.volread("f0_RFP.tif")

# get tracks and contours
tracks = pd.read_csv("Tracking_Results_H449.1_f0.csv") # tracking positions and the corresponding contours
contours = np.load("Contours_H449.1_f0.npy", allow_pickle=True)

contours32 = []
for c in contours:
    cont = []
    for pos in c:
        cont.append(pos.astype(np.float32))
    contours32.append(np.array(cont))
contours32 = np.array(contours32)
 
tracks["Contours"] = contours32

print(tracks.shape)
tracks.head()

# get predictions
predictions = pd.read_csv("Measures_H449.1_f0_1-689.csv", sep=",", index_col="idx")

print(predictions.shape)
predictions.head()
```

## For each mother and daughter cell, associate the corresponding contours with the predictions

Not all the cells have predictions : only the ones which were explicitly measured. For those cells, we want to append to the dataframe that contains the contours the corresponding TTA predictions.

TODO: The for loop is very long and it could be possible to optimize in some way using the pandas magic (choosing subsets, etc.) 

```python
# join the predictions to the tracks
count = 0
preds = []
for i, row in tracks.iterrows():
    if i % 10000 == 0:
        print(f"Line {i}")
        print(f"Number of retrieved cells: {count}")
    added_mother, added_daughter = False, False
    if not predictions[(predictions["mom_x"] == row["X"]) & (predictions["mom_y"] == row["Y"]) & (predictions["frame"] == row["Frame"])].empty: # retrieve the same mother cell
        preds.append(predictions[(predictions["mom_x"] == row["X"]) & (predictions["mom_y"] == row["Y"]) & (predictions["frame"] == row["Frame"])]["tta_pred"].values[0])
        count += 1
        added_mother = True
    if not predictions[(predictions["daugh_x"] == row["X"]) & (predictions["daugh_y"] == row["Y"]) & (predictions["frame"] == row["Frame"])].empty:  # daughter cell
        preds.append(predictions[(predictions["daugh_x"] == row["X"]) & (predictions["daugh_y"] == row["Y"]) & (predictions["frame"] == row["Frame"])]["tta_pred"].values[0])
        count += 1
        added_daughter = True
#     else:
#         print("missing daughter")
    if not added_mother and not added_daughter:  # if this cell is not in the mother or daughter cells that were measured
        preds.append(-1)

print(f"Number of retrieved cells : {count}")
tracks["predictions_tta"] = preds

tracks.head()
```

```python
# join the predictions to the tracks

k, count = 0, 0
contours = []  # contains couple of contours [mother_contour, daughter_contour]


for i, row in predictions.iterrows():
    c = {}  # [mother contour, daughter contour]
    if k % 100 == 0:
        print(f"Line {k}")
        print(f"Number of retrieved cells: {count}")

    if not tracks[(row["mom_x"] == tracks["X"]) & (row["mom_y"] == tracks["Y"]) & (row["frame"] == tracks["Frame"])].empty: # retrieve the same mother cell
#         print(row["frame"], row["mom_ID"], tracks[(row["mom_x"] == tracks["X"]) & (row["mom_y"] == tracks["Y"]) & (row["frame"] == tracks["Frame"])]["Contours"].values[0][0])
        c["mom_contours"] = tracks[(row["mom_x"] == tracks["X"]) & (row["mom_y"] == tracks["Y"]) & (row["frame"] == tracks["Frame"])]["Contours"].values[0]
        c["mom_ID"] = row["mom_ID"]
        c["frame"] = row["frame"]
        count += 1

    if not tracks[(row["daugh_x"] == tracks["X"]) & (row["daugh_y"] == tracks["Y"]) & (row["frame"] == tracks["Frame"])].empty:  # daughter cell
        c["daugh_contours"] = tracks[(row["daugh_x"] == tracks["X"]) & (row["daugh_y"] == tracks["Y"]) & (row["frame"] == tracks["Frame"])]["Contours"].values[0]
        c["daugh_ID"] = row["daugh_ID"]
        c["frame"] = row["frame"]
        count += 1

    if len(c) == 5:
#         print(row["frame"], row["mom_ID"], c["mom_contours"][0])
        contours.append(c)
    elif len(c) == 3:
        if "mom_contours" in c:
            c["daugh_contours"] = -1
            c["daugh_ID"] = -1
        else:
            c["mom_contours"] = -1
            c["mom_ID"] = -1

        contours.append(c)
    else: # no contours retrieved
        c["mom_contours"] = -1
        c["daugh_contours"] = -1
        c["mom_ID"] = -1
        c["daugh_ID"] = -1
        c["frame"] = -1
        contours.append(c)


    k += 1

contours = pd.DataFrame(contours)

print(f"Number of retrieved cells : {count}")

print(contours.shape)
predictions["mom_contours"] = contours["mom_contours"]
predictions["daugh_contours"] = contours["daugh_contours"]

predictions.head()
```

```python
# replace the missing contours
for i, c in contours.iterrows():
    if type(c["mom_contours"]) == int and c["mom_contours"] == -1:
        print("mom here")
        if (contours.loc[i - 1]["mom_ID"] == contours.loc[i]["mom_ID"]) or (contours.loc[i - 1]["daugh_ID"] == contours.loc[i]["daugh_ID"]):
            contours.at[i, "mom_contours"] = contours.loc[i - 1]["mom_contours"]
        else:
            contours.at[i, "mom_contours"] = contours.loc[i + 1]["mom_contours"]
    if type(c["daugh_contours"]) == int and c["daugh_contours"] == -1:
        print("daugh here")
        if (contours.loc[i - 1]["mom_ID"] == contours.loc[i]["mom_ID"]) or (contours.loc[i - 1]["daugh_ID"] == contours.loc[i]["daugh_ID"]):
            contours.at[i, "daugh_contours"] = contours.loc[i - 1]["daugh_contours"]
        else:
            k = 0
            while type(contours.loc[i+k]["daugh_contours"]) == int and contours.loc[i+k]["daugh_contours"] == -1:
                k += 1
            contours.at[i, "daugh_contours"] = contours.loc[i + k]["daugh_contours"]
```
