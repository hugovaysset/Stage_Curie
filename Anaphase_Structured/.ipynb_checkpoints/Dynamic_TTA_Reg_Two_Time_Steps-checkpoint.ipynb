# Dynamic prediction of TTA : $(X_{t-1}, X_t) \mapsto TTA_t$

First attempt to dynamic prediction which is direct regression of TTA but with the integration a temporal component.

```python
%matplotlib inline
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy
import seaborn as sns
from skimage import measure # to get contours from masks
import sklearn

import napari

sns.set_theme()
plt.rc('axes', labelsize=14, titlesize=15) 
```

## Loading Data

```python
os.chdir("D:\Documents\STAGE\Anaphase\Structured")
data = pd.read_csv("Measures_BFf0001_H449pos2_merged.csv", sep=";", comment="#", index_col="idx")
data = data[(data["time_to_anaphase"] >= 0)].dropna()

print(data.shape)
data.head()
```

# 1. Single feature : Daughter perimeter

```python
# prepare the training set
from sklearn.model_selection import train_test_split

features = ["daugh_per"]

for f in features:
    f_lag1 = []
    
    for idx, traj in data.groupby(level=0):   
        for i, time in enumerate(traj["time"]):
            if time == 0:
                f_lag1.append(np.nan)
            else:
                f_lag1.append(traj.iloc[i-1][f])
    data[f"{f}_lag1"] = f_lag1

for i in range(len(features)):
    features.append(f"{features[i]}_lag1")

data = data.dropna()

X_train, X_test, y_train, y_test = train_test_split(data[features].values, data["time_to_anaphase"].values, train_size=0.5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot(data[f"{features[0]}_lag1"], data[f"{features[0]}"], ".")
ax.set_xlabel("Daughter perimeter(t-1)")
ax.set_ylabel("Daughter perimeter(t)")
fig.tight_layout()
plt.show()
```

## Fitting and prediction

- Linear regression : R^2 = 0.77 and MSE on test set is 5.4 minutes.

### Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor

lm = RandomForestRegressor(max_depth=3, n_estimators=20).fit(X_train, y_train)
rmse = np.sqrt(np.mean((lm.predict(X_train) - y_train) ** 2))

print(f"R^2 coefficient on train set: {round(lm.score(X_train, y_train), 2)}")
print(f"RMSE on train set: {round(rmse, 2)} min.")

predictions_on_train_set = lm.predict(X_train)

plot = False
if plot:
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(X_train[:, 0], X_train[:, 1], y_train, c="b")
    ax.scatter(X_train[:, 0], X_train[:, 1], predictions_on_train_set, marker="+", c="r")
    
    ax.set_xlabel(features[0])
    ax.set_ylabel(features[1])
    ax.set_zlabel('TTA (min)')
    ax.set_title("Data distribution")
```

```python
if type(lm) == LinearRegression:
    print(f"Regression equation : {lm.coef_[0]} * {features[0]} + {lm.coef_[1]} * {features[1]} + {lm.intercept_}")
elif type(lm) in [RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor]:
    print(f"Features importance: {lm.feature_importances_}.")
```

### Evaluation

```python
predictions_on_test_set = lm.predict(X_test)
rmse = np.sqrt(np.mean((lm.predict(X_test) - y_test) ** 2))

print(f"R^2 coefficient on train set: {round(lm.score(X_test, y_test), 2)}")
print(f"RMSE on test set: {round(rmse, 2)} min.")

plot = False
if plot:
    %matplotlib qt
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(X_test[:, 0], X_test[:, 1], y_test, c="b")
    ax.scatter(X_test[:, 0], X_test[:, 1], predictions_on_test_set, marker="+", c="r")
    
    ax.set_xlabel(features[0])
    ax.set_ylabel(features[1])
    ax.set_zlabel('TTA (min)')
    ax.set_title("Prediction on test set.")
```

```python
%matplotlib inline
# evaluation with the clasical RMSE
losses, signed_losses = [], []
for obs, pred, gt in zip(X_test, predictions_on_test_set, y_test):
    loss = np.sqrt((pred - gt) ** 2)
    signed_losses.append(gt - pred)
    losses.append(loss)
losses = np.array(losses) 
signed_losses = np.array(signed_losses)

print(f"Mean absolute error on test set: {losses.mean()}")

fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# RMSE distribution
sns.histplot(losses, ax=axes[0])
axes[0].set_title("L2 loss distribution over all examples.")
axes[0].set_xlabel("L2 loss")

# Conditional distributions given the sign of the error
sns.histplot(losses[signed_losses >= 0], color="g", kde=True, label="Early predictions", ax=axes[1])
sns.histplot(losses[signed_losses < 0], color="r", kde=True, label="Late predictions", ax=axes[1])
axes[1].legend()
axes[1].set_xlabel("L2 loss")
axes[1].set_title("L2 loss given the sign of (y_true - y_pred)")
```

## 1.3 Repeated K-fold cross validation

```python
from sklearn.model_selection import RepeatedKFold

features = ["daugh_per", "daugh_per_lag1"]

X, y = data[features].values, data["time_to_anaphase"].values
n_folds, n_iter = 2, 100
kf = RepeatedKFold(n_splits=n_folds, n_repeats=n_iter)

rmse, scores = [], []
for train_idx, test_idx in kf.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    lm = LinearRegression().fit(X_train, y_train)
#     lm = RandomForestRegressor(max_depth=5, n_estimators=100).fit(X_train, y_train)
    scores.append(lm.score(X_test, y_test))
    rmse.append(np.sqrt(((y_test - lm.predict(X_test)) ** 2).mean()))
rmse = np.array(rmse)
scores = np.array(scores)

print(f"R^2 on {n_folds} folds repeated {n_iter} times : {scores}")
print(f"Mean R^2 on {n_folds}-fold cross-validation repeated {n_iter} times : {round(scores.mean(), 2)} +- {round(scores.std(), 2)}.\n")

print(f"RMSE on {n_folds} folds repeated {n_iter} times : {rmse}")
print(f"Mean RMSE on {n_folds}-fold cross-validation repeated {n_iter} times : {round(rmse.mean(), 2)}  +- {round(rmse.std(), 2)} min.")
```
