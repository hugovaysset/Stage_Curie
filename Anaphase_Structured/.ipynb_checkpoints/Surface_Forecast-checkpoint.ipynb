## Predicting $X_{t} = f(t)$

Here we want to predict the future values of the relative surface based on its past values. We want to have: $$\forall i, \forall t, f : t \mapsto X^i_{t}$$ by fitting this function on the early observations of the time series. This very simple approach doesn't involve any learning step since the curve is only fitted on the early observations of the time series (the five first ones here). This approach may be weak since the number of early points is really low and there might be a big number of possible curves that perfectly fit the e.g. three first points... This is where the learning step could help. But its strenght is that it can include time series of different lengths seemlessly (which isn't the case with classical ML models which need constant size inputs.

```python
# colors
import matplotlib.colors as mcolors
by_hsv = [(tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in mcolors.TABLEAU_COLORS.items()]
names = [name for hsv, name in by_hsv]

fig, ax = plt.subplots(1, 2, figsize=(14, 7))

idx = 15
subset = data.loc[idx].copy()
subset = subset.set_index("time")
subset = subset.drop(0)
subset["time_inv"] = 1 / subset.index.values
subset["relat_surf_inv"] = 1 / subset["relat_surf"]

print(subset.columns)

ax[0].plot(subset.index.unique(), subset["relat_surf"], c=names[idx % len(names)])
ax[0].scatter(subset[subset["anaphase"]].index, subset[subset["anaphase"]]["relat_surf"], c=names[idx % len(names)])

ax[1].plot(subset["time_inv"], subset["relat_surf_inv"])
ax[1].scatter(subset[subset["anaphase"]]["time_inv"], subset[subset["anaphase"]]["relat_surf_inv"], s=20, c=names[idx % len(names)])
    
ax[0].set_xlabel("time ")
ax[0].set_ylabel("relative surface")
ax[0].set_title("Relative surface over time")

ax[1].set_xlabel("1 / time ")
ax[1].set_ylabel("1 / relative surface")
ax[1].set_title("Inverse relative surface over inverse time")
# ax[1].set_ylim(0, 5)
```

## Fit a linear regression on the first observations

```python
from sklearn.linear_model import LinearRegression

X_train, y_train, X_test, y_test = subset[:3]["time_inv"].values.reshape(-1, 1), subset[:3]["relat_surf_inv"].values.reshape(-1, 1), subset[3:]["time_inv"].values.reshape(-1, 1), subset[3:]["relat_surf_inv"].values.reshape(-1, 1) 
lm = LinearRegression().fit(X_train, y_train)

print(f"R^2 coefficient : {round(lm.score(X_train, y_train), 2)}")

forecast = lm.predict(subset["time_inv"].values.reshape(-1, 1))

print(forecast.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

ax[0].plot(subset.index.unique(), subset["relat_surf"])
ax[0].plot(subset.index.unique(), 1 / forecast, "--", c="r")
ax[0].scatter(subset[subset["anaphase"]].index, subset[subset["anaphase"]]["relat_surf"], c="r")
ax[0].set_xlabel("time ")
ax[0].set_ylabel("relative surface")
ax[0].set_title("Relative surface over inverse time")

ax[1].plot(subset["time_inv"], subset["relat_surf_inv"])
ax[1].plot(subset["time_inv"], forecast, "--", c="r")
ax[1].scatter(subset[subset["anaphase"]]["time_inv"], subset[subset["anaphase"]]["relat_surf_inv"], c="r")
ax[1].set_xlabel("1 / time ")
ax[1].set_ylabel("1 / relative surface")
ax[1].set_title("Inverse relative surface over inverse time")
```

## Forecasting $X_{t+1} = f(X_{\leq t})$

Here we use lag features to learn a model that maps the current and past value of $X$ to its future value.

### Which model to choose ?

The linear regression model gives a MAE of 0.11 on test set.
SVR gives a MAE of 0.08 on test set.
Random Forest Regressor gives a MAE of 0.10.

The performances seem to improve when we taking more steps into account.
