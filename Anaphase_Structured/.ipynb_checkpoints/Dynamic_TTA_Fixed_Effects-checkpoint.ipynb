# TTA Regression with fixed effects

What I missed since two weeks was the good statistical terms for my literature research! I didn't find anything about "linear regression with correlated variables", the time series formalism didn't suit in my case (neither regarding the type of predictions that I wanted to make, nor the type of data : in TS they observe a single time series, often quite long and with a high temporal resolution, and we most often want to predict $X_{> t}$ from $X_{\leq t}$ vs. here we observe multiple "repeats" of the same temporal process over several individuals and I want to regress $y_t$ given $X_{\leq t}$), survival seemed to fit because of the individual-level predictions and the nature of the predictions but it wasn't suitable for time-varying covariates even people say it is...

In fact the terms that I missed were "panel data" and "fixed effect", which is common in biostatistics. Panel data are a specific case of longitudinal data : we repeat measurements on the same individual over time, this is exactly what I have. The problem with this data is that it violates the i.i.d residuals asusmptions made for classical statistical models like linear regression. To circumvent this, we decompose the residuals in two terms : a fixed, additive term $\alpha_i$, specific to one individual $i$ and a random gaussian noise term $\epsilon_{i,t}$. The model becomes : 
$$ y_{i,t} = X_{i,t}\beta + \alpha_i + \epsilon_{i,t} $$
This corresponds to allowing the intercept of the linear regression function to vary between individuals, this is exactly what I wanted to do! Let's do it in python. If we assume that $\forall i, \alpha_i$ is constant over time, then $\alpha_i - \bar{\alpha}_i = 0$ and then
$$ y_{i,t} - \bar{y}_{i} = (X_{i,t} - \bar{X}_i)\beta + 0 + (\epsilon_{i, t} - \bar{\epsilon}_{i})  $$

## 1. Tests for linear regression hypothesis violations

### 1.1 Fit a linear model (standard OLS) to get residuals

````python
exog = sm.tools.tools.add_constant(data['relat_surf'])
endog = data['time_to_anaphase']

mod = linearmodels.PooledOLS(endog, exog)
pooledOLS_res = mod.fit(cov_type='clustered', cluster_entity=True)
# Store values for checking homoskedasticity graphically

fittedvals_pooled_OLS = pooledOLS_res.predict().fitted_values
residuals_pooled_OLS = pooledOLS_res.resids
````

### 1.2 Homoskedasticity test

````python
# 3A. Homoskedasticity

# 3A.1 Residuals-Plot for growing Variance Detection
fig, ax = plt.subplots()
ax.scatter(fittedvals_pooled_OLS, residuals_pooled_OLS, color="blue")
ax.axhline(0, color = 'r', ls = '--')
ax.set_xlabel("Predicted", fontsize = 15)
ax.set_ylabel("Residuals", fontsize = 15)
ax.set_title("Homoskedasticity Test", fontsize = 15)
plt.show()

# 3A.2 White-Test
from statsmodels.stats.diagnostic import het_white, het_breuschpagan

print("Performing White test ...")

pooled_OLS_dataset = pd.concat([data, residuals_pooled_OLS], axis=1)
pooled_OLS_dataset = pooled_OLS_dataset.drop(["time"], axis = 1).fillna(0)
exog = sm.tools.tools.add_constant(data["relat_surf"]).fillna(0)
white_test_results = het_white(pooled_OLS_dataset["residual"], exog)
labels = ["LM-Stat", "LM p-val", "F-Stat", "F p-val"] 
print(dict(zip(labels, white_test_results)))# 3A.3 Breusch-Pagan-Test

breusch_pagan_test_results = het_breuschpagan(pooled_OLS_dataset["residual"], exog)
labels = ["LM-Stat", "LM p-val", "F-Stat", "F p-val"] 
print(dict(zip(labels, breusch_pagan_test_results)))

print("Homoscedasticity kind of respected.")
````

### Autocorrelation in residuals test

````python
# 3.B Non-Autocorrelation
# Durbin-Watson-Test
from statsmodels.stats.stattools import durbin_watson

durbin_watson_test_results = durbin_watson(pooled_OLS_dataset["residual"]) 
print(durbin_watson_test_results)

if 0 <= durbin_watson_test_results and durbin_watson_test_results < 1.5:
    print("Positive autocorrelation between residuals indicated.")
elif 1.5 <= durbin_watson_test_results < 2.5:
    print("No autocorrelation indicated.")
else:
    print("Negative autocorrelation detected.")
````

## 2. Build the Fixed-Effects model

````python
# FE und RE model
from linearmodels import PanelOLS
from linearmodels import RandomEffects

exog = sm.tools.tools.add_constant(data['relat_surf'])
endog = data["time_to_anaphase"]

# random effects model
model_re = RandomEffects(endog, exog) 
re_res = model_re.fit()

# fixed effects model
model_fe = PanelOLS(endog, exog, entity_effects = True) 
fe_res = model_fe.fit() 

#print results
print(re_res)
print(fe_res)
````

````python
import numpy.linalg as la
from scipy import stats
import numpy as np

def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B)) 
    pval = stats.chi2.sf(chi2, df)
    return chi2, df, pval

hausman_results = hausman(fe_res, re_res) 

print("chi-Squared: " + str(hausman_results[0]))
print("degrees of freedom: " + str(hausman_results[1]))
print("p-Value: " + str(hausman_results[2]))

if hausman_results[2] <= 0.05:
    print("H0 rejected, should use the Fixed effect model.")
````

## Evaluation of the fixed effect model

#### How to make predictions using the linear model ?
The ```fit``` method returns a ```PanelEffectsResults``` that contains all that we want (we call it ```fe_res``` in what follows).
1. The parameters (mean intercept and slope) are contained in ```fe_rs.params```.
2. The fixed (entity) effects are contained in ```fe_res.estimated_effects```. It is a dataframe with the same index as the training data. Be careful the fixed effect (intercept) is relative to the intercept that is mentionned in the model's fit summary... So to make good predictions, we should add this "mean" intercept with the ones that are reported in the estimated effects.

... We see that the predictions are really good!

### Plot the benefits of the linear FE model compared with the standard OLS

````python
idx = [15, 1, 43, 72, 0]
# idx = [0, 4]
colors = ["r", "b", "g", "orange", "black"]

intercept, slope = fe_res.params

fig, axes = plt.subplots(1, 2, figsize=(24, 8))

# OLS regression
ax0 = axes[0]
from sklearn.linear_model import LinearRegression
X, y = data["relat_surf"].values.reshape(-1, 1), data["time_to_anaphase"].values
lm = LinearRegression().fit(X, y)
ols_predictions = lm.predict(X)
rmse_ols = np.sqrt(((ols_predictions - y) ** 2).mean())
ax0.scatter(data["relat_surf"], data["time_to_anaphase"], c="b")
ax0.plot(X, ols_predictions, ":", c="grey", label=f"OLS, RMSE: {round(rmse_ols, 2)} min")
ax0.legend()
ax0.set_xlim(-0.01, 0.55)
ax0.set_title("Anonymized observations and standard OLS predictions")
ax0.set_xlabel("Relative surface")
ax0.set_ylabel("TTA (min)")

# Fixed effects regression
ax = axes[1]
for k, i in enumerate(idx):
    fixed_effect = fe_res.estimated_effects.loc[(i, 0)].values[0]
    predictions, gt = (intercept + fixed_effect) + data.loc[(i,)]["relat_surf"].values * slope, data.loc[(i,)]["time_to_anaphase"]
    loss = np.sqrt(((predictions - gt) ** 2).mean())
    ax.scatter(data.loc[(i,)]["relat_surf"], gt, marker=".", c=colors[k], label=f"RMSE {round(loss, 2)} min", s=60)
    ax.plot(data.loc[(i,)]["relat_surf"], predictions, c=colors[k])
ax.plot(X, ols_predictions, ":", c="grey", label=f"OLS, RMSE: {round(rmse_ols, 2)} min", alpha=0.5)
ax.legend()
ax.set_xlim(0, 0.55)
ax.set_title("Observations and fixed effect linear model predictions for four trajectories")
ax.set_xlabel("Relative surface")
ax.set_ylabel("TTA (min)")
````

### Plot the benefits the avering of the updated LFE predictions strategy

````python
# Evaluation with RMSE (of course this is artificial since we know)
intercept, slope = fe_res.params
    
avg_losses, losses = [], []
for idx, traj in data.groupby(level=0):
    fixed_effect = fe_res.estimated_effects.loc[(idx,)]["estimated_effects"].iloc[0]
    
    predictions, gt = (intercept + fixed_effect) + traj["relat_surf"].values * slope, traj["time_to_anaphase"].values
    averaged_predictions = np.array([np.mean([predictions[k] - 6 * (i-k-1) for k in range(i)]) for i in range(1, traj.shape[0] + 1)])  # we average the predictions made from the past until now to have a better estimate!
    for a in (averaged_predictions - gt) ** 2:
        avg_losses.append(a)
    for p in (predictions - gt) ** 2:
        losses.append(p)
avg_losses, losses = np.array(avg_losses), np.array(losses)

print(f"RMSE of the linear fixed effects model on the training dataset : {round(np.sqrt(losses.mean()), 2)} +- {round(np.sqrt(losses.std()), 2)} min.")
print(f"RMSE of the averaged linear fixed effects model on the training dataset : {round(np.sqrt(avg_losses.mean()), 2)} +- {round(np.sqrt(avg_losses.std()), 2)} min.")
````

## Out-of-sample predictions with the fixed effect model

Now we want to tackle the real issue for making out-of-sample predictions with this linear FE model : how to estimate the value of the intercept $\alpha_i$ when we observe a new individual ? How to estimate it accurately and as soon as possible ?

### 1. Can the fixed effect $\alpha_i$ be predicted given the first observations of $X_{i,t}$

We see that the value of the fixed effect (intercept) is not correlated with the initial surface value, explaining why we have trouble when trying to estimate it from early observations based on early observations.

````python
colors = ["r", "b", "g", "orange", "black", "yellow", "cyan", "magenta", "grey", "pink"]

fig, ax = plt.subplots(1, 1, figsize=(9, 9))
for idx, traj in data.groupby(level=0):
    ax.scatter(traj.loc[(slice(None), (6, 12, 18)),"relat_surf"].sum(), traj.loc[(slice(None), (6)), "fixed_effect"], c=colors[idx % len(colors)])
ax.set_xlabel("X_0")
ax.set_ylabel("Fixed_effect")
ax.set_title("Fixed effect vs. mean(X(t=18), X(t=24), X(t=30))")
````

### 2. Predicting the fixed effect for an out-of-sample observation with a static K-Neighbour estimation

We use a K-Nearest Neighbours approach to identify the training sample which is closest to the new observation. 500-fold cross-validation in this setting (KNN, K=5) gives a TTA RMSE of 6.14 +- 0.62 min which isn't than the classical predictions made from static regression or other approaches, but way under what is possible when "cheating" and estimating the fixed effect (we should consider this result as the best we could do). 

But one thing can be improved : we could average the past predictions to get an unbiased estimator of the true intercept. This would allow to improve our results : we get on 100-fold "cross-validation" (overlapping sampling) a mean RMSE of 5.3 +- 0.72 minutes.
This is not consistantly better than what was done before. Still far from what we would like to obtain (if possible).

### 1. Estimating $\alpha_i$ with a simple KNN-model

We observe that the predictions are really not well correlated with the true intercept values, which is consistent with the fact the intercept is not correlated with the early observations.

````python
os.chdir("D:/Hugo/Anaphase")
data = pd.read_csv("Analysis_BF_f0001-1-100.1.csv", sep=";", comment="#", 
                   index_col=["idx", "time"])
data = data[data["time_to_anaphase"] >= 0]
data = data[data["relat_surf"] > 0]

time = data.index.get_level_values("time").to_list()
data["time"] = pd.Categorical(time)

# train the FE lm to get the gt intercept for all
X, y = data["relat_surf"], data["time_to_anaphase"]
X = sm.tools.tools.add_constant(X)

from linearmodels import PanelOLS

# fixed effects model
model_fe = PanelOLS(y, X, entity_effects=True) 
fe_res = model_fe.fit() 

intercept, slope = fe_res.params
data["fixed_effect"] = fe_res.estimated_effects

data.head()
````

````python
# Now we train/evaluate a KNN model to predict the FE for out-of-sample observations
# Train test split
train_size = 0.75
index_values = data.index.get_level_values(0).unique()
X_train, y_train = data.loc[index_values[:int(train_size * index_values.size)]]["relat_surf"], data.loc[index_values[:int(train_size * index_values.size)]]["fixed_effect"]
X_test, y_test = data.loc[index_values[int(train_size * index_values.size):]]["relat_surf"], data.loc[index_values[int(train_size * index_values.size):]]["fixed_effect"]
y_test_tta = data.loc[index_values[int(train_size * index_values.size):]]["time_to_anaphase"]

X_train, X_test = X_train.values.reshape(-1, 1), X_test.values.reshape(-1, 1) 

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape, y_test_tta.shape)

# KKN Model to predict TTA from relat_surf
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor().fit(X_train, y_train)
fe_pred = knn.predict(X_test)

# store predictions and features in a dataframe for covenience
results = []
for x, gt, tta, pred, idx, time in zip(X_test[:, 0], y_test.values, y_test_tta.values, fe_pred, y_test.index.get_level_values(0), y_test.index.get_level_values(1)):
    results.append({"relat_surf": x, "fixed_effect": gt, "time_to_anaphase": tta, "fe_pred": pred, "idx": idx, "time": time})
results = pd.DataFrame(results).set_index(["idx", "time"])
results.head()
````

````python
# Fitted vs. GT
colors = ["r", "b", "g", "orange", "black"]

fig, ax = plt.subplots(1, 2, figsize=(20, 8))
plotted_traj = 0
for idx, traj in results.groupby(level=0):
    ax[0].scatter(traj.fixed_effect, traj.fe_pred, c=colors[idx % len(colors)])
    tta_pred = (intercept + traj.fe_pred) + traj.relat_surf * slope
    tta_pred_gt = (intercept + traj.fixed_effect) + traj.relat_surf * slope
    loss, loss_gt = np.sqrt(((tta_pred - traj.time_to_anaphase) ** 2).mean()), np.sqrt(((tta_pred_gt - traj.time_to_anaphase) ** 2).mean())
    if plotted_traj < 5:
        ax[1].scatter(traj.relat_surf, traj.time_to_anaphase, marker=".", c=colors[idx % len(colors)], s=60)
        ax[1].plot(traj.relat_surf, tta_pred, c=colors[idx % len(colors)], label=f"Estimated intercept, RMSE {round(loss, 2)} min")
        ax[1].plot(traj.relat_surf, tta_pred_gt, "--", c=colors[idx % len(colors)], label=f"True intercept, RMSE {round(loss_gt, 2)} min")
        plotted_traj += 1
ax[0].plot([-10, 10], [-10, 10], "--")
ax[0].set_xlabel("GT")
ax[0].set_ylabel("Prediction")
ax[0].set_title("Fitted vs. Ground Truth")
ax[1].set_xlabel("Relative Surface")
ax[1].set_ylabel("TTA (min)")
ax[1].set_title("TTA vs. Relative Surface")
ax[1].legend() 
````

````python
intercept, slope = fe_res.params

meta_losses_gt, meta_losses = [], []
# 10-Fold cross-validation
for fold in range(500):
    # Now we train/evaluate a KNN model to predict the FE for out-of-sample observations
    # Train test split
    index_values = np.array(data.index.get_level_values(0).unique())
    np.random.shuffle(index_values) 
    
    # train test split
    train_size = 0.8
    X_train, y_train = data.loc[index_values[:int(train_size * index_values.size)]]["relat_surf"], data.loc[index_values[:int(train_size * index_values.size)]]["fixed_effect"]
    X_test, y_test = data.loc[index_values[int(train_size * index_values.size):]]["relat_surf"], data.loc[index_values[int(train_size * index_values.size):]]["fixed_effect"]
    y_test_tta = data.loc[index_values[int(train_size * index_values.size):]]["time_to_anaphase"]

    X_train, X_test = X_train.values.reshape(-1, 1), X_test.values.reshape(-1, 1) 

    # KKN Model
    from sklearn.neighbors import KNeighborsRegressor

    knn = KNeighborsRegressor().fit(X_train, y_train)
    fe_pred = knn.predict(X_test)

    # store predictions and features in a dataframe for covenience
    results = []
    for x, gt, tta, pred, idx, time in zip(X_test[:, 0], y_test.values, y_test_tta.values, fe_pred, y_test.index.get_level_values(0), y_test.index.get_level_values(1)):
        results.append({"relat_surf": x, "fixed_effect": gt, "time_to_anaphase": tta, "fe_pred": pred, "idx": idx, "time": time})
    results = pd.DataFrame(results).set_index(["idx", "time"])

    losses_gt, losses = [], []
    # Fixed effects regression
    for idx, traj in results.groupby(level=0):
        tta_pred = (intercept + traj.fe_pred) + traj.relat_surf * slope
        tta_pred_gt = (intercept + traj.fixed_effect) + traj.relat_surf * slope
        loss, loss_gt = np.sqrt(((tta_pred - traj.time_to_anaphase) ** 2).mean()), np.sqrt(((tta_pred_gt - traj.time_to_anaphase) ** 2).mean())
        losses.append(loss)
        losses_gt.append(loss_gt)
    
    meta_losses.append(np.mean(np.array(losses)))
    meta_losses_gt.append(np.mean(np.array(losses_gt)))
    
meta_losses, meta_losses_gt = np.array(meta_losses), np.array(meta_losses_gt)

print(f"Mean RMSE on 5-fold CV for the positive reference (known alpha): {round(meta_losses_gt.mean(), 2)} +- {round(meta_losses_gt.std(), 2)} min.")
print(f"Mean RMSE on 5-fold CV with FE estimation with 5-NN (real oos prediction): {round(meta_losses.mean(), 2)} +- {round(meta_losses.std(), 2)} min.")
````

### 2. Averaging the model's prediction to get a better $\alpha_i$ estimate

We could do two things : averaging the predictions made by the KNN to get a single intercept value for all the observations of the same individual, or we could average the updated predictions made by the linear regression to estimate the TTA.

The first method gives 6.14 +- 0.4 min (100 folds) ; the second one gives 5.38 +- 0.43 min which seems better

````python
# Now we train/evaluate a KNN model to predict the FE for out-of-sample observations
# Train test split
train_size = 0.50
index_values = data.index.get_level_values(0).unique()
X_train, y_train = data.loc[index_values[:int(train_size * index_values.size)]]["relat_surf"], data.loc[index_values[:int(train_size * index_values.size)]]["fixed_effect"]
X_test, y_test = data.loc[index_values[int(train_size * index_values.size):]]["relat_surf"], data.loc[index_values[int(train_size * index_values.size):]]["fixed_effect"]
y_test_tta = data.loc[index_values[int(train_size * index_values.size):]]["time_to_anaphase"]

X_train, X_test = X_train.values.reshape(-1, 1), X_test.values.reshape(-1, 1) 

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape, y_test_tta.shape)

# KKN Model to predict TTA from relat_surf
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor().fit(X_train, y_train)
fe_pred = knn.predict(X_test)

# store predictions and features in a dataframe for covenience
results = []
for x, gt, tta, pred, idx, time in zip(X_test[:, 0], y_test.values, y_test_tta.values, fe_pred, y_test.index.get_level_values(0), y_test.index.get_level_values(1)):
    results.append({"relat_surf": x, "fixed_effect": gt, "time_to_anaphase": tta, "fe_pred": pred, "idx": idx, "time": time})
results = pd.DataFrame(results).set_index(["idx", "time"])
results.head()

# Averaging the updated past predictions of intercept among each individual
fe_pred_avgs = []
for idx, traj in results.groupby(level=0):
    predictions = traj["fe_pred"].values  # regular predictions
    fe_pred_avgs.extend([np.mean([predictions[k] for k in range(i)]) for i in range(1, traj.shape[0] + 1)])  # average updated past observations
results["fe_pred_avg"] = np.array(fe_pred_avgs)

# Fitted vs. GT
colors = ["r", "b", "g", "orange", "black"]

fig, ax = plt.subplots(1, 2, figsize=(20, 8))
plotted_traj = 0
for idx, traj in results.groupby(level=0):
    ax[0].scatter(traj.fixed_effect, traj.fe_pred_avg, c=colors[idx % len(colors)])
    predictions = ((intercept + traj.fe_pred_avg) + traj.relat_surf * slope).values  # regular predictions
    tta_pred = np.array([np.mean([predictions[k] - 6 * (i-k-1) for k in range(i)]) for i in range(1, traj.shape[0] + 1)])  # average updated past observations
    tta_pred_gt = (intercept + traj.fixed_effect) + traj.relat_surf * slope
    loss, loss_gt = np.sqrt(((tta_pred - traj.time_to_anaphase) ** 2).mean()), np.sqrt(((tta_pred_gt - traj.time_to_anaphase) ** 2).mean())
    if plotted_traj < 5:
        ax[1].scatter(traj.relat_surf, traj.time_to_anaphase, marker=".", c=colors[idx % len(colors)], s=60)
        ax[1].plot(traj.relat_surf, tta_pred, c=colors[idx % len(colors)], label=f"Estimated intercept, RMSE {round(loss, 2)} min")
        ax[1].plot(traj.relat_surf, tta_pred_gt, "--", c=colors[idx % len(colors)], label=f"True intercept, RMSE {round(loss_gt, 2)} min")
        plotted_traj += 1
ax[0].plot([-10, 10], [-10, 10], "--")
ax[0].set_xlabel("GT")
ax[0].set_ylabel("Prediction")
ax[0].set_title("Fitted vs. Ground Truth")
ax[1].set_xlabel("Relative Surface")
ax[1].set_ylabel("TTA (min)")
ax[1].set_title("TTA vs. Relative Surface")
ax[1].legend() 
````
