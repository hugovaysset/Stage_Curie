# Assessing within-across *individual* variability in the division process

```python
%matplotlib inline
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy
import seaborn as sns
from skimage import measure # to get contours from masks
import sklearn

import statsmodels.api as sm
import statsmodels.formula.api as smf

sns.set_theme()
plt.rc('axes', labelsize=20, titlesize=30) 
plt.rc("legend", fontsize=15)
plt.style.use('ggplot')
```

```python
os.chdir("D:\Hugo\Anaphase\Inter_Div_Correlation\H449.1")
ana_df = pd.read_csv("H449.1_anaphase_events.csv", sep=",")
ana_df.head()

# helper functions
def get_features_at_anaphase(ana_idx):
    """
    Return features of the queried division event (indexed by ana_idx) at the anaphase instant
    """
    return ana_df[(ana_df["ana_ID"] == ana_idx) & (ana_df["anaphase"])]

def get_related_events_features_at_anaphase(ana_idx):
    """
    Return features of all the division events involving the same cell as the queried cell (indexed by mom_idx)
    """
    mom_idx = ana_df[ana_df["ana_ID"] == ana_idx]["mom_ID"].unique()[0]
    return ana_df[(ana_df["mom_ID"] == mom_idx) & (ana_df["anaphase"])]

def get_features_at_all_anaphases(mom_idx):
    """
    Return features of all the division events involving the mother cell (indexed by mom_idx)
    """
    return ana_df[(ana_df["mom_ID"] == mom_idx) & (ana_df["anaphase"])]

def get_features_at_all_divisions(mom_idx):
    """
    Return features of all the division events involving the mother cell (indexed by mom_idx)
    """
    return ana_df[ana_df["mom_ID"] == mom_idx]
```

# 1. Visualise data

### Divisions duration

We see that there are some outliers, which are the very long division events. We can filter them since the typical duration should be around 30 minutes (between the apparition of the daughter cell and the beginning of the anaphase). 

```python
fig, ax  = plt.subplots(1, 1, figsize=(15, 6))

division_durations = []
for ana_idx in ana_df["ana_ID"].dropna().unique():
    event = ana_df[ana_df["ana_ID"] == ana_idx]
    t0, tf = event["time"].iloc[0], event[event["anaphase"]]["time"].values[0]
    division_durations.append(tf - t0)

ax.hist(division_durations)
ax.set_title("Division duration distribution (from daughter cell apparition to anaphase start)")
ax.set_xlabel("Time (min)")
ax.set_ylabel("Count")
```

```python
# Filter values
to_remove = []
for ana_idx in ana_df["ana_ID"].dropna().unique():
    event = ana_df[ana_df["ana_ID"] == ana_idx]
    t0, tf = event["time"].iloc[0], event[event["anaphase"]]["time"].values[0]
    if tf - t0 > 80 or tf - t0 <= 10:
        to_remove.extend(event.index.values)

ana_df = ana_df.drop(to_remove)
```

### Plot the distribution of number of divisions observed per individual

```python
ids = {k:0 for k in ana_df["mom_ID"].unique()}

several_div_idx = {}
for idx in ids:
    mom_occurences = ana_df[ana_df["mom_ID"] == idx]["ana_ID"].dropna().unique()
            
    ids[idx] = mom_occurences.size if mom_occurences.size > 0 else 0
    
    if mom_occurences.size > 2:
        several_div_idx[idx] = mom_occurences
#     if ids[idx] == 6:
#         print(idx, ana_df[ana_df["mom_ID"] == idx]["ana_ID"].unique())
    
counts = np.array(sorted([val for _, val in ids.items()]))

print(f"# individuals that divide two times or more : {np.sum(counts > 1)}")
print(f"# individuals that divide two times or more : {np.sum(counts > 2)}")

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.hist(sorted([val for _, val in ids.items()]), bins=8, align="mid")
ax.set_xlim(0.5, 8)
ax.set_title("Distribution of the number of cells given the number of divisions observed during one experiment")
```

### Plot the distribution of division features, grouped by individual

```python
division_durations = []
for mom_idx in several_div_idx.keys():
    
    for ana_idx in ana_df[ana_df["mom_ID"] == mom_idx]["ana_ID"].dropna().unique():
        event = ana_df[ana_df["ana_ID"] == ana_idx]
        t0, tf = event["time"].iloc[0], event[event["anaphase"]]["time"].values[0]
        
        final_per = event[event["anaphase"]]["daugh_per"].values[0]
        
#         if tf - t0 < 10 or tf - t0 >= 80:
#             continue
#         else:
        division_durations.append({"mom_ID": mom_idx, "ana_ID": ana_idx, "duration": tf - t0, "ana_daugh_per": final_per})
        
division_durations = pd.DataFrame(division_durations)
division_durations.shape
#     division_durations = at_anaphases["Frame"]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(30, 8))
fig.tight_layout()

# durations
ax = axes[0]
sns.boxplot(x="mom_ID", y="duration", data=division_durations, ax=ax)

medians = division_durations.groupby(['mom_ID'])['duration'].median().values
nobs = division_durations['mom_ID'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n=" + i for i in nobs]

overall_mean = division_durations["duration"].mean()
ax.axhline(overall_mean, c="red", linestyle="--", label="Overall mean")
 
# Add it to the plot
pos = range(len(nobs))
for tick ,label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick],
            medians[tick] + 0.03,
            nobs[tick],
            horizontalalignment='center',
            size='large',
            color='w',
            weight='semibold')
ax.legend()
ax.set_title("Division durations per individual")

# durations
ax = axes[1]
sns.boxplot(x="mom_ID", y="ana_daugh_per", data=division_durations, ax=ax)

medians = division_durations.groupby(['mom_ID'])['ana_daugh_per'].median().values
nobs = division_durations['mom_ID'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n=" + i for i in nobs]
 
# Add it to the plot
pos = range(len(nobs))
for tick ,label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick],
            medians[tick] + 0.03,
            nobs[tick],
            horizontalalignment='center',
            size='medium',
            color='w',
            weight='semibold')
ax.set_title("Daughter perimeter at anaphase, per individual")

overall_mean = division_durations["ana_daugh_per"].mean()
ax.axhline(overall_mean, c="red", linestyle="--", label="Overall mean")
ax.legend()
```

# 2. One-way ANOVA to test its assumptions

One-way ANOVA makes those three assumptions:
* Residuals are normally distributed -> ANOVA and then plotting of the residuals (Q-Q plot) and Shapiro-Wilk test
* Equal variance between conditions -> ANOVA and then plotting of residuals vs. fitted + Bartlett test
* Independance of samples (in particular, the successive measurements regarding the sequential divisions of the same individual should be independant from each other) -> 

### Perform ANOVA and plot the residuals

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Ordinary Least Squares (OLS) model
model = ols('duration ~ C(mom_ID)', data=division_durations).fit()
# print(model.params)
anova_table = anova_lm(model, typ=2)

# model residuals
division_durations["lm_residual"] = model.resid
symbols = ['D', '^']
colors = ['r', 'g', 'blue', "orange", "brown", "purple"]

fig, ax = plt.subplots(1, 1, figsize=(20, 8))

for k, mom_idx in enumerate(division_durations["mom_ID"].unique()):
    traj = division_durations[division_durations["mom_ID"] == mom_idx]
    ax.scatter([k] * traj.shape[0], traj["lm_residual"], color=colors[k % len(colors)])
ax.set_xlabel('Individual');
ax.set_ylabel('Residuals');
# ax.set_xticks(division_durations["mom_ID"].unique())
ax.set_title("Residuals of the linear model Division_Duration ~ Individual")
```

### 1. Normality test

We see that without the big outliers which are much longer than expected, the distribution of the residuals is quite nicely close to a normal distribution. 
- On the left of the Q-Q plot : residuals are above the line -> less values than expected here, i.e. less very small values (expected since the division duration is never under ~ 15 minutes) so short left tail.
- At the right of the Q-Q plot : residuals are above the line -> values increase more rapidly than in the standard gaussian, so long right tail.

Shapiro-Wilk test : (H0) : the data is normally distributed.

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

standard_res = (division_durations.lm_residual - division_durations.lm_residual.mean()) / division_durations.lm_residual.std()

w, pvalue = scipy.stats.shapiro(model.resid)
print(f"Shapiro-Wilk test statistics: W={w}, p-value={pvalue}.")

ax = axes[0]
sm.qqplot(standard_res, line='45', ax=ax)
ax.set_xlabel("Theoretical Quantiles")
ax.set_ylabel("Standardized Residuals")
ax.set_title("Q-Q plot of the ANOVA residuals")

# histogram
ax = axes[1]
ax.hist(model.resid, bins='auto', histtype='bar', ec='k', label=f"Shapiro-Wilk statistics : W={round(w, 4)}, p-value={round(pvalue, 4)}") 
ax.set_xlabel("Residuals")
ax.set_ylabel('Count')
ax.set_title("Distribution of the ANOVA residuals")

ax.legend()
```

### 2.Homoscedasticity test
Those two tests suggest that populations have equal variance... Even for the extreme cases (mom_ID == 22, very high variance and mom_ID == 37, very low variance), no test rejectcs the null hypothesis ((H0) : same variance).

Again, if we leave the outliers in the dataset, we will very likely reject the null hypothesis.

So...

```python
data = division_durations.set_index("mom_ID")["duration"]

w, pvalue = scipy.stats.bartlett(*[data.loc[i].values for i in data.index.unique()])
# w, pvalue = scipy.stats.bartlett(data.loc[22].values, data.loc[37].values)
print(f"Bartlett test: Statistic={w}, p-value={pvalue}")

w, pvalue = scipy.stats.levene(*[data.loc[i].values for i in data.index.unique()])
# w, pvalue = scipy.stats.levene(data.loc[22].values, data.loc[37].values)
print(f"Levene test: Statistic={w}, p-value={pvalue}")

w, pvalue = scipy.stats.fligner(*[data.loc[i].values for i in data.index.unique()])
# w, pvalue = scipy.stats.fligner(data.loc[22].values, data.loc[37].values)
print(f"Fligner test: Statistic={w}, p-value={pvalue}")
```

### 3. Independance test

Check if the successive divisions are independant from each other, or if there is some historical influence

We show visually that there is a mild or none effect of time on the division duration and on the division duration or on the final daughter cell perimeter. We thus conclude that we can perform ANOVA-like tests by grouping the successive divisions per individual, without taking the temporal aspect into account. Other tests on the slope of each curve could be made to render this a bit more rigorous.

```python
division_durations = []
for mom_idx in several_div_idx.keys():
    
    for i, ana_idx in enumerate(ana_df[ana_df["mom_ID"] == mom_idx]["ana_ID"].dropna().unique()):
        event = ana_df[ana_df["ana_ID"] == ana_idx]
        t0, tf = event["time"].iloc[0], event[event["anaphase"]]["time"].values[0]
        final_per = event[event["anaphase"]]["daugh_per"].values[0]
        
        division_durations.append({"mom_ID": mom_idx, "ana_ID": ana_idx, "duration": tf - t0, "replicate": i+1, "ana_daugh_per": final_per})
        
division_durations = pd.DataFrame(division_durations)

fig, axes = plt.subplots(1, 2, figsize=(30, 8))

ax = axes[0]
for mom_idx in division_durations["mom_ID"]:
    traj = division_durations[division_durations["mom_ID"] == mom_idx]
    ax.plot(traj["replicate"], traj["duration"])
    
mean = division_durations.groupby("replicate")["duration"].mean()
ax.plot(division_durations["replicate"].unique(), mean, "--", label="average duration")

ax.set_xlim(0.5, 5.5)
ax.set_title("Division duration as a function of the division position in each cell's divisions sequence")
ax.set_xlabel("Division index (position in the sequence)")
ax.set_ylabel("Division duration (min)")
ax.legend()

ax = axes[1]
for mom_idx in division_durations["mom_ID"]:
    traj = division_durations[division_durations["mom_ID"] == mom_idx]
    ax.plot(traj["replicate"], traj["ana_daugh_per"])
    
mean = division_durations.groupby("replicate")["ana_daugh_per"].mean()
ax.plot(division_durations["replicate"].unique(), mean, "--", label="average final daughter perimeter")

ax.set_xlim(0.5, 5.5)
ax.set_title("Final daughter cell perimeter as a function of the division position in each cell's divisions sequence")
ax.set_xlabel("Division index (position in the sequence)")
ax.set_ylabel("Final daughter cell perimeter (min)")
ax.legend()
```

# 3. One-way ANOVA

The ANOVA suggests that there is no signficant within-individual effect on the division duration or on the final daughter cell perimeter.

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Ordinary Least Squares (OLS) model for division duration
model = ols('duration ~ C(mom_ID)', data=division_durations).fit()
anova_table = anova_lm(model, typ=2)

print("ANOVA results for the model Division_duration ~ Individual.")
print(anova_table)

# Ordinary Least Squares (OLS) model for final daughter cell perimeter 
model = ols('ana_daugh_per ~ C(mom_ID)', data=division_durations).fit()
anova_table = anova_lm(model, typ=2)

print("\n\nANOVA results for the model Final_daughter_perimeter ~ Individual.")
print(anova_table)
```
