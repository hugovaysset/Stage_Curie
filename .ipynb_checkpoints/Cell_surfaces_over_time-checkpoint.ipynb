# Analyze the cells' surface over time during division

```python
import matplotlib.pyplot as plt
import napari
import numpy as np
import pandas as pd
import seaborn as sns

%gui qt5
```

```python
data = pd.read_csv("D:/Documents/STAGE/Anaphase/surfaces_over_time.csv", sep=";", index_col="idx")
print(data.head())
print(data.shape)
```

```python
# add relative surfaces (daughter / mom) column
rs = []
for idx, line in data.iterrows():
    rs.append(line["daugh_surf"] / line["mom_surf"])

data["relat_surf"] = rs
print(data.head())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

for idx in data.index.unique():
    event = data.loc[idx]
    ax.plot(event["time"], event["relat_surf"], label=idx)
    
# plot anaphase box
ax.plot([0, 55], [0.5, 0.5], "--", c="grey")
ax.plot([0, 55], [0.55, 0.55], "--", c="grey")
ax.plot([45, 45], [0., 0.6], "--", c="grey")
ax.plot([55, 55], [0., 0.6], "--", c="grey")

ax.set_xlabel("time (min)")
ax.set_ylabel("relative daughter/mom surface")

ax.set_title("Relative daughter cell / mom cell over time")
```
